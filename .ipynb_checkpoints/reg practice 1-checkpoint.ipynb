```python
import requests
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
response = requests.get('https://www.quandl.com/api/v3/datasets/YAHOO/INDEX_DJI.json').json()
```

```python
data = pd.DataFrame(response['dataset']['data'], columns=response['dataset']['column_names'])
```

```python
data['Date'] = pd.to_datetime(data['Date'], errors='raise')
```

```python
data = data.set_index('Date').sort_index()
```

```python
data['prev_close'] = data['Close'].shift(1)
data['SMA5'] = data['prev_close'].rolling(window=5).mean()
```

```python
data.dropna(inplace=True)
```

```python
len(data)
```

```python
reg = LinearRegression()
```

```python
split = int(data.shape[0] * .4)
```

```python
test = data.ix[split:]
```

```python
train = data.ix[:split]
```

```python
reg.fit(train[['prev_close','SMA5']], train['Close'])
```

```python
predictions = reg.predict(test[['prev_close','SMA5']])
```

```python
test['prediction'] = predictions
```

```python
mse = mean_squared_error(predictions, test['Close'])
```

```python
mse
```

```python
plt.plot(test.index, test['Close'], color='g')
```

```python
plt.plot(test.index, test['prediction'], color='r')
plt.show()
```
